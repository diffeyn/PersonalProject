```python
import pandas as pd
import re
```

```python
matches = pd.read_csv('../../data_files/scraped_raw/matches/matches_past.csv')
```

```python
matches.head()
```

```python
month_map = {
        "jan": "01", "feb": "02", "mar": "03", "apr": "04",
        "may": "05", "jun": "06", "jul": "07", "aug": "08",
        "sep": "09", "oct": "10", "nov": "11", "dec": "12"
    }

matches['date'] = matches['date'].str.replace(r"^[A-Za-z]+ ", "", regex=True)

matches['date'] = matches['date'].str.lower()


matches['date'] = matches['date'].str.replace(",", "", regex=False)

matches['date'] = matches['date'].apply(
        lambda x: re.sub(
            r"([A-Za-z]+)",
            lambda m: month_map[m.group(1).lower()],
            x
        )
    )

matches['date'] = matches['date'].apply(lambda x: x + " 2025" if re.search(r"\d{4}$", x) is None else x)

matches['date'] = pd.to_datetime(matches['date'], format="%m %d %Y")
```

```python
### subset matches to only after 2023-02-01

matches_merge = matches[matches['date'] >= '2023-02-01'].copy()

matches_merge
```

```python
### if away or home team is KCW, change to SKC

matches_merge['home_team'] = matches_merge['home_team'].replace('KCW', 'SKC')
matches_merge['away_team'] = matches_merge['away_team'].replace('KCW', 'SKC')
```

```python
## find unqiue team names in each dataframe

players_teams = players['club'].unique().tolist()

team_teams = pd.unique(team[['home_team', 'away_team']].values.ravel()).tolist()

master_teams = pd.unique(master[['home_team', 'away_team']].values.ravel()).tolist()

feed_teams = pd.unique(feed[['home_team', 'away_team']].values.ravel()).tolist()

print("Players teams:", players_teams)
print("Team teams:", team_teams)
print("Master teams:", master_teams)
print("Feed teams:", feed_teams)
```

```python
shortname_fix = {
    "LAG": "LA",
    "DCU": "DC",
    None: None,
    "nan": None
}
```

```python
def normalize_short(short):
    if pd.isna(short):
        return None

    s = str(short).strip().upper()

    # fix known mismatches
    if s in shortname_fix:
        return shortname_fix[s]

    return s
```

```python
players['team_clean'] = players['club'].apply(normalize_short)
team['team_clean_home'] = team['home_team'].apply(normalize_short)
team['team_clean_away'] = team['away_team'].apply(normalize_short)
master['team_clean_home'] = master['home_team'].apply(normalize_short)
master['team_clean_away'] = master['away_team'].apply(normalize_short)
feed['team_clean_home'] = feed['home_team'].apply(normalize_short)
feed['team_clean_away'] = feed['away_team'].apply(normalize_short)
```

```python
## find unqiue team names in each dataframe

players_teams = players['club'].unique().tolist()

team_teams = pd.unique(team[['home_team', 'away_team']].values.ravel()).tolist()

master_teams = pd.unique(master[['home_team', 'away_team']].values.ravel()).tolist()

feed_teams = pd.unique(feed[['home_team', 'away_team']].values.ravel()).tolist()

print("Players teams:", players_teams)
print("Team teams:", team_teams)
print("Master teams:", master_teams)
print("Feed teams:", feed_teams)
```

```python
master_long = pd.concat([
    master[["match_id", "date", "home_team"]]
        .rename(columns={"home_team": "team"})
        .assign(side="home"),

    master[["match_id", "date", "away_team"]]
        .rename(columns={"away_team": "team"})
        .assign(side="away")
])
```

```python
players_idl = players.merge(
    master_long,
    on=["date", "team", "side"],
    how="left"
)
```

```python
### match feed id 

feed_matched = feed.merge(
    master[["date", "home_team", "away_team", "match_id"]],
    on=["date", "home_team", "away_team"],
    how="left",
    suffixes=("", "_master")
)

feed_matched.head()


## replace match_id in feed with match_id_master

feed = feed_matched.drop(columns=['match_id'])

feed = feed.rename(columns={'match_id_master': 'match_id'})

feed
```
