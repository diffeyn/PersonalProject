```python
import pandas as pd

```

```python
feed = pd.read_csv("../../data_files/scraped_raw/matches/feed/mls_match_feeds.csv")

feed_2 = feed.copy()

# Normalize title column (lowercase, strip)
feed_2["title"] = feed_2["title"].astype(str).str.lower().str.strip()

# Drop meaningless events
meaningless = ["kick off", "kickoff", "end of second half", "end of first half"]
feed_2 = feed_2[~feed_2["title"].isin(meaningless)]
# Drop match_id entirely
feed_no_id = feed_2.drop(columns=["match_id"], errors="ignore")

# Now find exact duplicates based on useful columns
dupes = feed_no_id[feed_no_id.duplicated(keep=False)]

# Sort for human readability
dupes = dupes.sort_values(["date", "minute", "comment"])

dupes = pd.DataFrame(dupes)

dupes
```

```python
### drop second duplicates from original feed

feed_cleaned = feed.drop(dupes.index[::2])

### find rows count in original and cleaned feeds

print(f"Original feed rows: {len(feed)}")
print(f"Cleaned feed rows: {len(feed_cleaned)}")
```

```python
feed_cleaned.to_csv("../../data_files/scraped_raw/matches/feed/mls_match_feeds_no_dupes.csv", index=False)
```

```python
team = pd.read_csv("../../data_files/scraped_raw/matches/teams/team_stats_combined.csv")

team_no_id = team.drop(columns=["match_id"], errors="ignore")

dupe_team_rows = team_no_id[team_no_id.duplicated(keep=False)]

dupe_team_rows = pd.DataFrame(dupe_team_rows)

dupe_team_rows
```

```python
### players

players = pd.read_csv("../../data_files/scraped_raw/matches/players/mls_match_player_stats.csv")
```

```python
players_no_id = players.drop(columns=["match_id"], errors="ignore")

dupe_player_rows = players_no_id[players_no_id.duplicated(keep=False)]

dupe_player_rows = pd.DataFrame(dupe_player_rows)

dupe_player_rows
```

```python
### drop second duplicates from original team stats

players_dropped = players.drop(dupe_player_rows.index[::2])

## save cleaned team stats

players_dropped.to_csv("../../data_files/scraped_raw/matches/players/mls_match_player_stats_no_dupes.csv", index=False)
```
